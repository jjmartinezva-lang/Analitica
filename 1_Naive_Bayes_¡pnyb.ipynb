 **Modelo de Pre Aprobación**

Caso de estudio: Se requiere implementar un modelo de pre aprobación de créditos a partir de las variables socio económicas de un solicitante de crédito.

* Edad: Indica el número de años de un solicitante de crédito
* Ingresos: Indica todos los ingresos que recibe una persona
* Egresos: Indica todos los hastos mensuales de una persona
*Monto (EAD) : Es el monto que una persona está solicitando a través del crédito

La variable de referencia será la variable PreAprobación. Esta variable indica la pre aprobacion o pre negación de un crédito.

```python
import numpy as np
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
```

2. Cargamos Datos de trabajo

```python
nxl = '/content/sample_data/6. SolicitantesCrédito(USD).xlsx'
XDB = pd.read_excel(nxl, sheet_name=0)

#Sekeccionamos las variables de trabajo
XDB=XDB[["Edad", "Ingresos", "Egresos","Monto (EAD)" , "PreApr"]]
XDB=XDB.dropna(subset= ["PreApr"])
display(XDB.head())

#Separar variables de entrada - variables de salida
XD=XDB[["Edad","Ingresos","Egresos","Monto (EAD)"]] #Entrada
yd=XDB["PreApr"] #Variable de salida
```

3. Se implementa el modelo de clasificación

```python
mnb=GaussianNB() #Coco de modelo
mnb.fit(XD,yd) #Entrenamos el modelo

#Datos por categoría
ndat=mnb.class_count_
print("El número de datos por categoría es: \n",ndat)
cat=mnb.classes_
print("Las categorías son: \n",cat)

#Determinamos la propiedad estadistica de las variables
u=mnb.theta_
print("La mededida de las variables es : \n:",u)
sigma=np.sqrt(mnb.var_)
print("La desviación de las variables es: \n", sigma)
print("Los límites superiores de las variables son \n" , u+sigma)
print("Los limites inferiores de las variables son: \n", u-sigma)
```
